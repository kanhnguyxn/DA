Đối với decimal thì chúng ta cần dựa vào symbol để điền giá trị phù hợp
decimal của token đề cập đến đơn vị nhỏ nhất mà mã thông báo tiền điện tử có thể được chia. Nó cho biết mã thông báo có thể có bao nhiêu vị trí thập phân khi được giao dịch hoặc đại diện. Số thập phân là một phần thông số kỹ thuật của mã thông báo và được xác định trong hợp đồng thông minh của nó. Sau khi được đặt, số thập phân của mã thông báo phải không thay đổi, vì việc thay đổi giá trị thập phân có thể dẫn đến sự không nhất quán và lỗi trong giao dịch mã thông báo
WETH có decimal là 18, ETH có decimal là 18,..

```python
# với symbol là WETH và ETH thì decimal là 18, còn USDC thì decimal là 6
df_collection_data['decimals'] = df_collection_data['symbol'].apply(lambda x: 18 if x in ['WETH', 'ETH'] else 6)

df_collection_data.groupby('symbol')[['decimals', 'symbol']].count()
```

```python
# chuyển đổi kiểu dữ liệu
# Sử dụng dtypes để chuyển đổi kiểu dữ liệu
df_collection_data.dtypes
# cột price phải là float và event_type là kiểu datetime
```

```python
# chuyển đổi định dạng cột time
df_collection_data['event_timestamp'] = pd.to_datetime(df_collection_data['event_timestamp'], unit='s')
# chuyển đổi kiểu dữ liệu cột payment quantity
df_collection_data['price'] = df_collection_data['price'].astype('float64')
```

```python
df_collection_data.dtypes
```

```python

# chuyển giá trị trong cột price về cùng 1 giá trị tiền tệ sẽ lad ETH
# 1WETH = 0,9896 ETH.
# 1USDC = 0,00031ETH
# tính giá trị của cột price bằng decimal*price*
def convert_price(row):
    convert_price = row['price'] / (10 ** row['decimals'])
    if row['symbol'] == 'WETH':
        return round(convert_price * 0.9896, 4)
    elif row['symbol'] == 'USDC':
        return round(convert_price * 0.00031, 4)
    else:
      return convert_price  

# áp dụng cho từng dòng
df_collection_data['price'] = df_collection_data.apply(convert_price, axis=1)
```

```python
# lưu dữ liệu vào sql
import pyodbc
import os
from dotenv import load_dotenv
# thông tin server
load_dotenv()

server = os.getenv('SERVER')
database = os.getenv('DATABASE')

if server is None:
    print("Environment variable 'SERVER' is not set.")
if database is None:
    print("Environment variable 'DATABASE' is not set.")
# thông tin server
load_dotenv()

server = os.getenv('SERVER')
database = os.getenv('DATABASE')

if server is None:
    print("Environment variable 'SERVER' is not set.")
if database is None:
    print("Environment variable 'DATABASE' is not set.")
try:
    # tạo kết nối windown authentication
    connection = pyodbc.connect(
        'DRIVER={ODBC Driver 17 for SQL Server};SERVER=' + server + 
        ';DATABASE=' + database + 
        ';Trusted_Connection=yes;'
    )
    cursor = connection.cursor()

    # tên table trong sql
    transaction_table_name = 'transaction_infor'
    nfts_info_table_name = 'nfts_infor'

    # kiểm tra bảng đã tồn tại chưa
    create_transaction_table_query = f'''
    IF NOT EXISTS (SELECT * FROM sysobjects WHERE name='{transaction_table_name}' AND xtype='U')
    CREATE TABLE {transaction_table_name} (
        name VARCHAR(50),
        [transaction] VARCHAR(200),
        seller VARCHAR(50),
        buyer VARCHAR(50),
        event_type VARCHAR(10),
        chain VARCHAR(10),
        quantity INT,
        price FLOAT,
        decimals INT,
        symbol VARCHAR(10),
        time DATETIME,
        FOREIGN KEY (name) REFERENCES {nfts_info_table_name}(name)
    )
    '''

    cursor.execute(create_transaction_table_query)
    connection.commit()

    # câu lệnh insert dữ liệu
    merge_query = f'''
    MERGE INTO {transaction_table_name} AS target
    USING (VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)) AS source (name, [transaction], seller, buyer, event_type, chain, quantity, price, decimals, symbol, time)
    ON target.name = source.name
       AND target.[transaction] = source.[transaction]
       AND target.seller = source.seller
       AND target.buyer = source.buyer
       AND target.event_type = source.event_type
       AND target.chain = source.chain
       AND target.quantity = source.quantity
       AND target.price = source.price
       AND target.decimals = source.decimals
       AND target.symbol = source.symbol
       AND target.time = source.time
    WHEN MATCHED THEN 
        UPDATE SET name = source.name, [transaction] = source.[transaction], seller = source.seller, buyer = source.buyer, event_type = source.event_type, chain = source.chain, quantity = source.quantity, price = source.price, decimals = source.decimals, symbol = source.symbol, time = source.time
    WHEN NOT MATCHED THEN
        INSERT (name, [transaction], seller, buyer, event_type, chain, quantity, price, decimals, symbol, time)
        VALUES (source.name, source.[transaction], source.seller, source.buyer, source.event_type, source.chain, source.quantity, source.price, source.decimals, source.symbol, source.time);
    '''

    # insert dữ liệu
    for index, row in df_collection_data.iterrows():
        cursor.execute(merge_query, row['name'], row['transaction'], row['seller'], row['buyer'], row['event_type'], row['chain'], row['quantity'], row['price'], row['decimals'], row['symbol'], row['event_timestamp'])
        connection.commit()

    # đóng kết nối
    cursor.close()
    connection.close()

except pyodbc.Error as ex:
    print("An error occurred while connecting to the database:", ex)
```
